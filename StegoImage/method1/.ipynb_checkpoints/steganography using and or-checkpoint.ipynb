```python
import tensorflow as tf
import matplotlib.pyplot as plt
from numpy import *

# Load the cover image and secret image
cover_image = tf.io.read_file("cover_image.jpg")
secret_image = tf.io.read_file("secret_image.jpg")

# Decode the images
cover_image = tf.image.decode_jpeg(cover_image)
secret_image = tf.image.decode_jpeg(secret_image)

# Resize the secret image to the same size as the cover image
secret_image = tf.image.resize(secret_image, tf.shape(cover_image)[:2])

# Normalize the images
cover_image = tf.image.per_image_standardization(cover_image)
secret_image = tf.image.per_image_standardization(secret_image)

# Create the stego image by combining the cover image and secret image
# Split the cover image into its RGB channels
r, g, b = tf.split(cover_image, 3, axis=-1)
# Split the secret image into its RGB channels
secret_r, secret_g, secret_b = tf.split(secret_image, 3, axis=-1)
```

```python
r=array(r)
r=r.astype(int)

g=array(g)
g=g.astype(int)

b=array(b)
b=b.astype(int)
```

```python
from numpy import *
sr=array(secret_r)
sr=sr.astype(int)

sg=array(secret_g)
sg=sg.astype(int)

sb=array(secret_b)
sb=sb.astype(int)

```

```python
# Combine the RGB channels of the cover image and secret image
stego_r = array(tf.bitwise.bitwise_or(r, sr))
stego_g = array(tf.bitwise.bitwise_or(g, sg))
stego_b = array(tf.bitwise.bitwise_or(b, sb))
stego_image = array(tf.concat([stego_r, stego_g, stego_b], axis=-1))

# Encode the stego image and save it
stego_image = array(tf.image.encode_jpeg(stego_image))

tf.io.write_file("stego_image.jpg", stego_image)
```

```python
# To extract the secret image
# Load the stego image
stego_image = tf.io.read_file("stego_image.jpg")
stego_image = tf.image.decode_jpeg(stego_image)
stego_image = tf.image.per_image_standardization(stego_image)
```

```python
# Split the stego image into its RGB channels
stego_r, stego_g, stego_b = tf.split(stego_image, 3, axis=-1)

stego_r=array(stego_r)
stego_r=stego_r.astype(int)

stego_g=array(stego_g)
stego_g=stego_g.astype(int)

stego_b=array(stego_b)
stego_b=stego_b.astype(int)
```

```python
secret_r=sr
secret_g=sg
secret_b=sb
```

```python
# Extract the secret image from the stego image by separating the RGB channels
secret_r = array(tf.bitwise.bitwise_and(stego_r, secret_r))
secret_g =array( tf.bitwise.bitwise_and(stego_g, secret_g))
secret_b = array(tf.bitwise.bitwise_and(stego_b, secret_b))

secret_image = array(tf.concat([secret_r, secret_g, secret_b], axis=-1))

cover_r = array(tf.bitwise.bitwise_and(stego_r, r))
cover_g =array( tf.bitwise.bitwise_and(stego_g, g))
cover_b = array(tf.bitwise.bitwise_and(stego_b, b))

cover_image = array(tf.concat([cover_r, cover_g, cover_b], axis=-1))
```

```python
#Normalize the secret image
secret_image = array(tf.image.per_image_standardization(secret_image))
cover_image = array(tf.image.per_image_standardization(cover_image))
```
